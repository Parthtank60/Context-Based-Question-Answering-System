```
print('Loading word vectors...')
word2vec = {}
with open('/content/drive/My Drive/glove.6B.100d.txt') as f:
  for line in f:
    values = line.split()
    word = values[0]
    vec = np.asarray(values[1:], dtype='float32')
    word2vec[word] = vec
print('Found %s word vectors.' % len(word2vec))
```

```
print('Filling pre-trained embeddings...')
num_words = min(MAX_NUM_WORDS, len(word2idx) + 1)
embedding_matrix = np.zeros((num_words, EMBEDDING_DIM))
print(embedding_matrix.shape)
for word, i in word2idx.items():
  if i < num_words:
    embedding_vector = word2vec.get(word)
    if embedding_vector is not None:
      embedding_matrix[i] = embedding_vector
```

```
#create embedding layer
embedding_layer = Embedding(
  num_words,
  EMBEDDING_DIM,
  weights=[embedding_matrix],
  input_length=max_context_len,
)

```

```
##### build the model #####
encoder_context_inputs_placeholder = Input(shape=(max_context_len,))
print("Shape of input context:",encoder_context_inputs_placeholder.shape)
x_context = embedding_layer(encoder_context_inputs_placeholder)
print("Shape of input context after embedding:",x_context.shape)
encoder_context1 = LSTM(
  LATENT_DIM,
  return_sequences=True,
  return_state=True,
)
encoder_context_outputs1, h_context1, c_context1 = encoder_context1(x_context)

print(encoder_context_outputs1.shape)
print(h_context1.shape)
print(c_context1.shape)


encoder_context2 = LSTM(
  LATENT_DIM,
  return_state=True,
)

encoder_context_outputs2, h_context2, c_context2 = encoder_context2(encoder_context_outputs1)

print(encoder_context_outputs2.shape)
print(h_context2.shape)
print(c_context2.shape)
```

```
encoder_question_inputs_placeholder = Input(shape=(max_len_question,))
print("Shape of input context:",encoder_question_inputs_placeholder.shape)

embedding_layer_question = Embedding(
  num_words,
  EMBEDDING_DIM,
  weights=[embedding_matrix],
  input_length=max_len_question,
)

x_question = embedding_layer_question(encoder_question_inputs_placeholder)
print("Shape of input context:",x_question.shape)
encoder_question1 = LSTM(
  LATENT_DIM,
  return_sequences=True,
  return_state=True,
)
encoder_question_outputs1, h_question1, c_question1 = encoder_question1(x_question)

print(encoder_question_outputs1.shape)
print(h_question1.shape)
print(c_question1.shape)

encoder_question2 = LSTM(
  LATENT_DIM,
  return_state=True,
)
encoder_question_outputs2, h_question2, c_question2 = encoder_question2(encoder_question_outputs1)

print(encoder_question_outputs2.shape)
print(h_question2.shape)
print(c_question2.shape)
```

```
from keras.layers import concatenate
h = concatenate([h_context2, h_question2])
c = concatenate([c_context2, c_question2])

print(h.shape)
print(h.shape)
encoder_states = [h, c]
```

```
LATENT_DIM_DECODER = 512
decoder_inputs_placeholder = Input(shape=(max_len_target,))
print("decoder input shape:",decoder_inputs_placeholder.shape)

decoder_embedding = Embedding(num_words, EMBEDDING_DIM, weights=[embedding_matrix])
decoder_inputs_x = decoder_embedding(decoder_inputs_placeholder)
print("decoder embedded input",decoder_inputs_x.shape)


decoder_lstm = LSTM(
  LATENT_DIM_DECODER,
  return_sequences=True,
  return_state=True,
)
decoder_outputs, _, _ = decoder_lstm(decoder_inputs_x, initial_state=encoder_states)
print(decoder_outputs.shape)
```

```
#final dense layer for predictions
decoder_dense = Dense(num_words, activation='softmax')

decoder_outputs = decoder_dense(decoder_outputs)
print(decoder_outputs.shape)
```

```

model = Model([encoder_context_inputs_placeholder, encoder_question_inputs_placeholder, decoder_inputs_placeholder], decoder_outputs)
```

```
from keras.optimizers import Adadelta
model.compile(
            optimizer='adam',
            loss='sparse_categorical_crossentropy',
            metrics=['accuracy'])
```

```
# map indexes back into real words
# so we can view the results
idx2word_eng = {v:k for k, v in word2idx.items()}
idx2word_trans = {v:k for k, v in word2idx.items()}
```

```
def decode_sequence(input_context, input_question):
  # Encode the input as state vectors.
  states_value = encoder_model.predict([input_context,input_question])
  
  # Generate empty target sequence of length 1.
  target_seq = np.zeros((1, 1))
  # Populate the first character of target sequence with the start character.
  # NOTE: tokenizer lower-cases all words
  target_seq[0, 0] = word2idx['sos']
  # if we get this we break
  eos = word2idx['eos']
  # Create the translation
  output_sentence = []
  for _ in range(max_len_target):
    output_tokens, h, c = decoder_model.predict(
      [target_seq] + states_value
    )


    # Get next word
    idx = np.argmax(output_tokens[0, 0, :])

    # End sentence of EOS
    if eos == idx:
      break

    word = ''
    if idx > 0:
      word = idx2word_trans[idx]
      output_sentence.append(word)

    # Update the decoder input
    # which is just the word just generated
    target_seq[0, 0] = idx

    # Update states
    states_value = [h, c]

  return ' '.join(output_sentence)
```
